# Plot results


```python
import pandas as pd


gender_inference_random_forest = pd.read_csv('./gender_inference_random-forest.csv', header=None)
gender_inference_logistic_regression = pd.read_csv('./gender_inference_logistic-regression.csv', header=None)
age_inference_random_forest = pd.read_csv('./age_inference_random-forest.csv', header=None)
age_inference_logistic_regression = pd.read_csv('./age_inference_logistic-regression.csv', header=None)



gender_inference_random_forest.rename(columns={0: 'dataset', 1: 'balanced acc', 2: 'f1', 3: ' precision', 4: ' recall', 5: 'roc auc'}, inplace=True)
gender_inference_logistic_regression.rename(columns={0: 'dataset', 1: 'balanced acc', 2: 'f1', 3: 'precision', 4: 'recall', 5: 'roc auc'}, inplace=True)
age_inference_random_forest.rename(columns={0: 'dataset', 1: 'RF accuracy'}, inplace=True)
age_inference_logistic_regression.rename(columns={0: 'dataset', 1: 'LR accuracy'}, inplace=True)
```

### Gender inference

```python
import matplotlib.pyplot as plt


plt.style.use('ggplot')
gender_inference_random_forest.plot(kind='bar')
indices = range(gender_inference_random_forest.shape[0])
labels = gender_inference_random_forest['dataset'].to_list()
plt.xticks(indices, labels)
plt.ylabel('Random Forest metrics')
plt.xlabel('Dataset')
```

```python
plt.style.use('ggplot')
gender_inference_logistic_regression.plot(kind='bar')
indices = range(gender_inference_logistic_regression.shape[0])
labels = gender_inference_logistic_regression['dataset'].to_list()
plt.xticks(indices, labels)
plt.ylabel('Logistic Regression metrics')
plt.xlabel('Dataset')
```

### Age category inference

```python
joined_df = pd.merge(age_inference_random_forest, age_inference_logistic_regression, on='dataset')


plt.style.use('ggplot')
joined_df.plot(kind='bar')
indices = range(joined_df.shape[0])
labels = joined_df['dataset'].to_list()
plt.xticks(indices, labels)
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
```
