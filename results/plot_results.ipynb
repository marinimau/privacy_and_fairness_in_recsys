# Plot results


```python
import pandas as pd


gender_inference_random_forest = pd.read_csv('./gender_inference_random_forest.csv', header=None)
gender_inference_logistic_regression = pd.read_csv('./gender_inference_logistic_regression.csv', header=None)
age_inference_random_forest = pd.read_csv('./age_inference_random_forest.csv', header=None)
age_inference_logistic_regression = pd.read_csv('./age_inference_logistic_regression.csv', header=None)

gender_inference_random_forest.rename(columns={0: 'dataset', 1: 'Random Forest'}, inplace=True)
gender_inference_logistic_regression.rename(columns={0: 'dataset', 1: 'Logistic Regression'}, inplace=True)
age_inference_random_forest.rename(columns={0: 'dataset', 1: 'Random Forest'}, inplace=True)
age_inference_logistic_regression.rename(columns={0: 'dataset', 1: 'Logistic Regression'}, inplace=True)

```

```python
gender_inference_random_forest.set_index('dataset')
gender_inference_logistic_regression.set_index('dataset')
age_inference_random_forest.set_index('dataset')
age_inference_logistic_regression.set_index('dataset')
```

## Gender inference


```python
import matplotlib.pyplot as plt




joined_df = pd.merge(gender_inference_random_forest, gender_inference_logistic_regression, on='dataset')


plt.style.use('ggplot')
joined_df.plot(kind='bar')
indices = range(joined_df.shape[0])
labels = joined_df['dataset'].to_list()
plt.xticks(indices, labels)
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
```

Age category inference

```python
joined_df = pd.merge(age_inference_random_forest, age_inference_logistic_regression, on='dataset')


plt.style.use('ggplot')
joined_df.plot(kind='bar')
indices = range(joined_df.shape[0])
labels = joined_df['dataset'].to_list()
plt.xticks(indices, labels)
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
```
