Plot nDCG



```python
import pandas as pd
import matplotlib.pyplot as plt


original = [pd.read_csv('./recommender-metrics/obfuscated/original/cutoff-10.csv', header=0),
            pd.read_csv('./recommender-metrics/obfuscated/original/cutoff-20.csv', header=0),
            pd.read_csv('./recommender-metrics/obfuscated/original/cutoff-50.csv', header=0)]

pred = [pd.read_csv('./recommender-metrics/obfuscated/pred/cutoff-10.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/pred/cutoff-20.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/pred/cutoff-50.csv', header=0)]

avg = [pd.read_csv('./recommender-metrics/obfuscated/avg/cutoff-10.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/avg/cutoff-20.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/avg/cutoff-50.csv', header=0)]

filtered = [pd.read_csv('./recommender-metrics/obfuscated/filtered/cutoff-10.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/filtered/cutoff-20.csv', header=0),
        pd.read_csv('./recommender-metrics/obfuscated/filtered/cutoff-50.csv', header=0)]
```

* Different cutoff, same model

```python
cutoff_list = [10, 20, 50]
models_name = ['bprmf', 'multidae', 'user_knn', 'wrmf']

def plot_different_cutoff(data, model_index):
    plt.figure()
    plt.style.use('ggplot')
    for d in data:
        ndcg_dict = {'nDCG': [d[0].loc[model_index, 'nDCG'], d[1].loc[model_index, 'nDCG'],
                         d[2].loc[model_index, 'nDCG']]}
        df = pd.DataFrame(data=ndcg_dict)
        plt.plot(df)
    plt.title('Model = ' + str(models_name[model_index]))
    indices = range(len(cutoff_list))
    labels = cutoff_list
    plt.xticks(indices, labels)
    plt.ylabel('nDCG')
    plt.xlabel('Model')
    legend = ['original', 'PerBlur pred', 'PreBlur avg', 'filtered']
    plt.legend(legend, loc='lower right')
```

```python
for i in range(len(models_name)):
    plot_different_cutoff([original, pred, avg, filtered], i)


```
