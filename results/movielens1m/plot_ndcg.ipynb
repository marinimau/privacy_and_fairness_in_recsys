Plot nDCG



```python
import pandas as pd
import matplotlib.pyplot as plt



cutoff_10 = pd.read_csv('./recommender-metrics/cutoff-10.csv', header=0)
cutoff_20 = pd.read_csv('./recommender-metrics/cutoff-20.csv', header=0)
cutoff_50 = pd.read_csv('./recommender-metrics/cutoff-50.csv', header=0)

cutoff_10_obfuscated = pd.read_csv('./recommender-metrics/obfuscated/cutoff-10.csv', header=0)
cutoff_20_obfuscated = pd.read_csv('./recommender-metrics/obfuscated/cutoff-20.csv', header=0)
cutoff_50_obfuscated = pd.read_csv('./recommender-metrics/obfuscated/cutoff-50.csv', header=0)

cutoff_50.head(8)
```

* Different cutoff, same model

```python
def get_cutoff_for_model(index, obfuscated=False):
    if obfuscated:
        data = {'nDCG': [cutoff_10_obfuscated.loc[index, 'nDCG'], cutoff_20_obfuscated.loc[index, 'nDCG'],
                         cutoff_50_obfuscated.loc[index, 'nDCG']]}
    else:
        data = {'nDCG': [cutoff_10.loc[index, 'nDCG'], cutoff_20.loc[index, 'nDCG'], cutoff_50.loc[index, 'nDCG']]}
    return pd.DataFrame(data=data)

cutoff_bprmf = get_cutoff_for_model(0)
cutoff_item_knn = get_cutoff_for_model(1)
cutoff_most_pop = get_cutoff_for_model(2)
cutoff_user_knn = get_cutoff_for_model(3)
cutoff_lightgcn = get_cutoff_for_model(4)
cutoff_multidae = get_cutoff_for_model(5)
cutoff_neumf = get_cutoff_for_model(6)
cutoff_wrmf = get_cutoff_for_model(7)

cutoff_bprmf_obfuscated = get_cutoff_for_model(0, obfuscated=True)
cutoff_user_knn_obfuscated = get_cutoff_for_model(2, obfuscated=True)
cutoff_multidae_obfuscated = get_cutoff_for_model(1, obfuscated=True)
cutoff_wrmf_obfuscated = get_cutoff_for_model(3, obfuscated=True)

```

```python
cutoff_list = [10, 20, 50]

def plot_different_cutoff(data, model):
    plt.figure()
    plt.style.use('ggplot')
    data.plot(kind='line')
    plt.title('Model = ' + str(model))
    indices = range(len(cutoff_list))
    labels = cutoff_list
    plt.xticks(indices, labels)
    plt.ylabel('nDCG')
    plt.xlabel('Model')
```

```python
plot_different_cutoff(cutoff_bprmf, 'BPRMF')
```

```python
plot_different_cutoff(cutoff_item_knn, 'item-KNN')
```

```python
plot_different_cutoff(cutoff_user_knn, 'user-KNN')
```

```python
plot_different_cutoff(cutoff_most_pop, 'Most Popular')
```

```python
plot_different_cutoff(cutoff_lightgcn, 'LightGCN')
```

```python
plot_different_cutoff(cutoff_multidae, 'MultiDae')
```

```python
plot_different_cutoff(cutoff_neumf, 'NeuMF')
```

```python
plot_different_cutoff(cutoff_wrmf, 'WRMF')
```

* Different model, same cutoff

```python
def plot_different_model(data, cutoff=10):
    plt.figure()
    plt.style.use('ggplot')
    data.plot(kind='bar')
    plt.title('Cutoff = ' + str(cutoff))
    indices = range(data.shape[0])
    labels = data['model'].to_list()
    plt.xticks(indices, labels)
    plt.ylabel('nDCG')
    plt.xlabel('Model')
```

```python
plot_different_model(cutoff_10, 10)
```

```python
plot_different_model(cutoff_20, 20)
```

```python
plot_different_model(cutoff_50, 50)
```

* obfuscated

```python
plot_different_model(cutoff_10_obfuscated, 10)
```
